資料合併:
因為有共通欄位Accident_Index，所以用此欄位合併
但因為兩個資料開始的時間點不一樣，所以只併擁有
共通欄位的資料。

```python
df_merge = pd.merge(data1,data2,on ="Accident_Index")
display(data1.head(),data2.head(),df_merge.head())
```
